```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

### x-y axis

For a function y = a*x + b, we random choose an integer value for a and b respectively.

```python
a = np.random.randint(2,10)
b = np.random.randint(2,10)
x = np.linspace(-2,10,100)
y = a*x+b
print('y = %d*x + %d'%(a,b))
plt.plot(x, y)
def func(x):
    return a*x + b
scatter = {}
for i in range(3):
    tmp = np.random.randint(10)
    while str(tmp) in scatter.keys():
        tmp = np.random.randint(10)
    scatter[str(tmp)] = func(tmp)

colors = ['red', 'green', 'blue']
i = 0
plt.xlabel("x")
plt.ylabel("y")
for k,v in scatter.items():
    plt.scatter(int(k),v,color=colors[i])
    print('x%d:%d, y%d:%d'%(i, int(k), i, v))
    i+=1
```

### Transfer x-y axis to a-b axis 

For the three point we random selected above, we transfer them to a-b axis, which will transfer them from a point to a line in the new space. You will find that the transferred lines will intersect at (a,b).

```python
t_a = np.linspace(-10,10,100)
i = 0
plt.xlabel("a")
plt.ylabel("b")
for k,v in scatter.items():
    t_b = int(k) * -t_a + v
    plt.plot(t_a,t_b,color=colors[i])
    i+=1
plt.scatter(a,b,marker='D',color='black')
```

In a similar way, if we transfer lines (y = a*x + b) intersect at a point from x-y axis to a-b axis, we will find these transferred points lie on the same line. Hough transfrom is an algorithm using similar idea, it transfers the points to curves in a new axisspectively. If there is any significant intersection in the new axis indicates there exists a line in the original space.
