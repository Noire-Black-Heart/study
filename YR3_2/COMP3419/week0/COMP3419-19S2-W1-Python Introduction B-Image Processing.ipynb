## Read Image in Python

```python
# import function[imread] from package[imageio]
from imageio import imread
```

```python
# read image from path specified and save it as variable[img]
# please change [img_path] to the path of the image you want to read 
# (absolute file path and relative file path, are both accepted)
# e.g. if you put the image 'foo.jpg' in the same folder of this file
# img = imread('./foo.jpg')
img_path = './example_usyd.jpg' # path to a image
img = imread(img_path)
```

## Plot Image

```python
# you need to install matplotlib library first
# "as" works as "rename"
from matplotlib import pyplot as plt
```

```python
# this line is for Jupyter to display image within this webpage
# It can be ignored if you are running py script from terminal
%matplotlib inline  

plt.imshow(img)
plt.show()
```

## Matrix Shape

```python
# 3D matrix/ array
# 3D => (Height, Width, ColorChannels_RGB)
print(img.shape)
```

## Access Pixel Element in Matrix

```python
# Fetch element stored at 0th row and 5th column => There should be 3 values(r,g,b)
img[0][5]
```

## Slice Operation for Matrix

```python
# Fetch the b_channel of patch 
# where its row-index is from 100 to 150 and column-index is from 200 to 250
crop_area = img[100:150, 200:250, 2]
print(crop_area)
```

```python
plt.imshow(crop_area, cmap='gray') # cmap, stands for, color_map. It is used for visualization purpose only.
```

## Colorful Image is a Matrix of 3 Channels [RGB]

```python
# ":" with no boundaries specified => no requirement/ constraint put on this dimension 
# we want to keep the image size unchanged, but only fetch different channels of it
r_channel = img[:,:,0]
g_channel = img[:,:,1]
b_channel = img[:,:,2]
```

```python
# Each slice (color channel) of a 3D Matrix is a 2D matrix
print(r_channel.shape)
print(g_channel.shape)
print(b_channel.shape)
```

```python
# Those 2D Matrix themselves do not contain any colorful(RGB) information
plt.imshow(g_channel, cmap='gray')
plt.show()
```
